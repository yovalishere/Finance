### Real Time Google 1day 1 min data with selection button (Candle stick)

```python
data = yf.download("GOOG", period='1d', interval='1m')
fig=go.Figure() # declare fig

########################  Candlestick  ############################
# Plot
fig.add_trace(go.Candlestick(x=data.index,# data.index=datetime
                open=data['Open'],
                high=data['High'],
                low=data['Low'],
                close=data['Close'], name = 'market data'))
# Title
fig.update_layout(
    title={'text':'<b>Google Live Share Price</b>',
            'yanchor': 'top'},
    yaxis_title='Stock Price (USD per Shares)')

# X-Axes (with button options)
fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=15, label="15m", step="minute", stepmode="backward"),
            dict(count=45, label="45m", step="minute", stepmode="backward"),
            dict(count=1, label="HTD", step="hour", stepmode="todate"),
            dict(count=3, label="3h", step="hour", stepmode="backward"),
            dict(step="all")
        ])
    )
)

#Show
fig.show()
```
